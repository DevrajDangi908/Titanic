```python
# *
# *
# *
# *
# *
# *
# *
# *
# *
# *
# *
```

```python
# Add A Title Column :-
```

```python
both_df['Title'] = both_df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
both_df.head(2)
```

```python
# Add A Family Size Column :- 
```

```python
both_df['FamilySize'] = both_df['SibSp'] + both_df['Parch'] + 1
both_df.head(2)
```

```python
# Add Deck Column :-
```

```python
both_df['Deck'] = both_df['Cabin'].astype(str).str[0]
both_df['Deck'] = both_df['Deck'].replace('n', 'Unknown') 
both_df.head(2)
```

```python
# Add Age Grouup column :-
```

```python

def get_passenger_type(age):
    if pd.isnull(age):
        return 'Unknown' 
    elif age < 12:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Old'
    
both_df['Age_Group'] = both_df['Age'].apply(get_passenger_type)
```

```python
both_df.head()
```

```python
# Add Column Fare - Per - Person :- 
```

```python
both_df['Fare_Per_Person'] = both_df['Fare'] / both_df['FamilySize']
```

```python
both_df.head(2)
```

```python
# Add Column Ticket Type :-
```

```python
both_df['Ticket_Type'] = both_df['Ticket'].apply(lambda x: x.split()[0].replace('.', '').replace('/', '').strip() if not x.isdigit() else 'Numeric')
```

```python
both_df.head(2)
```

```python
# these column missing value > 0.05 
```

```python
cols = [var for var in both_df.columns if both_df[var].isnull().mean() >0.05 ]
cols
```

```python
# Survived  mean  And Median :- 

survived_mean = both_df['Survived'].mean()
survived_mean
```

```python
survived_median = both_df['Survived'].median()
survived_median
```

```python
# Fill_Survived  mean  & add New Columns :-

both_df['fill_Survived_mean'] = both_df['Survived'].fillna(survived_mean)
```

```python
# Fill_Survived  median  & add New Columns :- 

both_df['fill_Survived_median'] = both_df['Survived'].fillna(survived_median)
```

```python
both_df.head()
```

```python
# Variance Summary :- 
```

```python
print(" Variance Summary ")
print("Original Survived col variance :- ",both_df['Survived'].var())
print("*******")
print("Fill Mean Survived Col Variance :- ",both_df['fill_Survived_mean'].var())
print("********")
print("fill Median Survived col Variance :- ",both_df['fill_Survived_median'].var())
```

```python
# Plot   :-
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
both_df['Survived'].plot(kind='kde',ax=ax)
both_df['fill_Survived_mean'].plot(kind='kde',ax=ax,color='red')
both_df['fill_Survived_median'].plot(kind='kde',ax=ax,color='green')
lines,labels = ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```
